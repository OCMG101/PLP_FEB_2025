## Task 3: Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Line graph for average selling price per brand
avg_prices = df.groupby('Brand')['Selling Price'].mean().sort_index()

plt.figure(figsize=(10,6))
plt.plot(avg_prices.index, avg_prices.values, marker='o')
plt.title('Average Selling Price by Phone Brand')
plt.xlabel('Brand')
plt.ylabel('Average Selling Price')
plt.xticks(rotation=60)
plt.grid(True)
plt
```

```python
(df.groupby("Brand")["Selling Price"].mean().sort_values().plot(kind="bar")
.set_title("Bar Chart Comparing the Average Price of Different Brands"))
```

```python
# Histogram
sns.histplot(df["Selling Price"], bins=30, kde=True).set_title("Histogram Showing the Distribution of Selling Price, Original Price, and Rating")
```
